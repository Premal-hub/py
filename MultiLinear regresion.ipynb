```python
import pandas as pd 
```

```python
data = {
    "area": [7420, 8960, 9960, 7500, 7420, 7500, 8580, 16200, 8100, 5750,
             13200, 6000, 6550, 3500, 7800, 6000, 6600, 8500, 4600, 6420],
    
    "bedrooms": [4,4,3,4,4,3,4,5,4,3,3,4,4,4,3,4,4,3,3,3],
    
    "price": [13300000,12250000,12250000,12215000,11410000,10850000,10150000,
              10150000,9870000,9800000,9800000,9681000,9310000,9240000,
              9240000,9100000,9100000,8960000,8890000,8855000]
}
df=pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[["area","bedrooms"]]
y=df[["price"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)

model.coef_
model.intercept_
```

```python
y_pre=model.predict(x_test)
y_pre
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print("R2_score",r2_score(y_test,y_pre))
print("MAS",mean_absolute_error(y_test,y_pre))
print("MSE",mean_squared_error(y_test,y_pre))
```

```python
# Predict price for home area=8000, bedrooms=4
new_data = pd.DataFrame({
    "area": [8000],
    "bedrooms": [4]
})

pred_price = model.predict(new_data)
print("Predicted Price =", pred_price[0])
```
