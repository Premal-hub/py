```python
import pandas as pd
```

```python
data = {
    'area': [2600, 3000, 3200, 3600, 4000, 2200, 2800, 3100, 3300, 3500,
             2500, 2700, 3400, 3800, 3900, 2950, 3050, 3150, 3250, 3450],
    
    'price': [550000, 565000, 610000, 680000, 725000, 480000, 540000, 590000,
              630000, 660000, 510000, 545000, 615000, 700000, 715000, 
              560000, 575000, 600000, 620000, 650000]
}
df=pd.DataFrame(data)
#df=pd.read_csv('homeprices.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='*')
```

```python
x=df[['area']]
y=df[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predictian=lr.predict(x_test)
y_predictian
```

```python
plt.scatter(x,y,c="blue")
plt.plot(x_train,lr.predict(x_train),c="red")
plt.xlabel("area")
plt.ylabel("prices")
plt.title("price is area ")
plt.show()

```

```python
#Supervise Matric
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("MAE" ,mean_absolute_error(y_test,y_predictian))
```

```python
print("MSE" ,mean_squared_error(y_test,y_predictian))
```

```python
import numpy as np
print("RMSE ",np.sqrt(mean_squared_error(y_test,y_predictian)))
```

```python
print("R2 score ", r2_score(y_test,y_predictian))
```

```python
r2=r2_score(y_test,y_predictian)
print("Adjested R2 socre ", 1-((1-r2)*(40-1)/(40-1-1)))
```
