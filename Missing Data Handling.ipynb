```python
import pandas as pd
import numpy as np
```

```python
data = {
    'id': list(range(1, 21)),
    'age': [24, 27, np.nan, 22, 25, 29, 31, np.nan, 23, 26,
            np.nan, 30, 28, 22, np.nan, 27, 25, 26, np.nan, 24],
    
    'salary': [45000, 54000, 50000, np.nan, 61000, 58000, np.nan,
               47000, 46000, np.nan, 52000, 60000, 58000,
               np.nan, 49000, 53000, np.nan, 52000, 56000, 48000],
    
    'city': ["Mumbai", "Delhi", "Bangalore", "Mumbai", np.nan, "Delhi", "Pune",
             "Delhi", "Mumbai", "Hyderabad", "Pune", "Mumbai", np.nan,
             "Chennai", "Kolkata", "Pune", "Mumbai", "Delhi", "Bangalore", np.nan]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
print("Missing vale Count: ",df.isnull().sum())
```

```python
df_remaove=df.dropna()
df_remaove
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
df['city'] = df['city'].fillna(df['city'].mode()[0])
```
