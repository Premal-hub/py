```python
# Creating dataset (No imports needed)
data = {
    "city": ["Mumbai", "Delhi", "Pune", "Mumbai", "Delhi"],
    "gender": ["Male", "Female", "Female", "Male", "Male"],
    "education": ["High School", "Graduate", "Post Graduate", "Graduate", "High School"],
    "salary": [50000, 42000, 60000, 52000, 45000]      # Target variable
}

import pandas as pd     # Only when needed
df = pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()

df['city']=le.fit_transform(df['city'])

df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(drop='first',sparse_output=False)

encode_gender=ohe.fit_transform(df[['gender']])

gender_df=pd.DataFrame(encode_gender,columns=ohe.get_feature_names_out(['gender']))

df= pd.concat([df,gender_df],axis=1)
df.drop("gender", axis=1, inplace=True)    
df
```

```python

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()  

df["education"] = oe.fit_transform(df[["education"]])

df


```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

df["salary"]=scaler.fit_transform(df[['salary']])
df
```
